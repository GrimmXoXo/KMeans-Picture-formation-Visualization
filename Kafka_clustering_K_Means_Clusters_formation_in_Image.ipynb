```python
import PIL#Library used to make an image into array or visa versa
import numpy as np
```

```python
kafka=np.asarray(PIL.Image.open('kafka.jpg'))#HSR kafka
```

```python
kafka.shape
```

```python
Image=kafka.reshape(-1,3)
```

```python
from IPython.display import display
PIL.Image.fromarray(kafka)#Original image
```

```python
k_clusters=[]
inertias=[]
from sklearn.cluster import KMeans
for k in range(1,11):
  kmeans=KMeans(n_clusters=k,random_state=42,n_init=10)
  kmeans.fit(Image)
  segmented_image=kmeans.cluster_centers_[kmeans.labels_].reshape(kafka.shape)
  display(PIL.Image.fromarray(segmented_image.astype(np.uint8)))#Can just use utf8 dont need np i think
  k_clusters.append(k)
  inertias.append(kmeans.inertia_)# See the process of image clusters forming
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.plot(k_clusters,inertias,marker='o',label='Inertia',color='b')
plt.xticks(k_clusters)
plt.grid(True)
plt.show()
```
